```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('s3://ia340-sakyitamai/Popular_Baby_Names.csv')

# Group by 'Child's First Name' and 'Year of Birth' and calculate the sum of 'Count'
grouped_df = df.groupby(['Child\'s First Name', 'Year of Birth'])['Count'].sum().reset_index()

# Sort the dataframe by 'Child's First Name' and 'Year of Birth'
grouped_df = grouped_df.sort_values(['Child\'s First Name', 'Year of Birth'])

# Calculate the growth rate
grouped_df['Growth Rate'] = grouped_df.groupby('Child\'s First Name')['Count'].pct_change()

# Pivot the dataframe for the line chart
pivot_df = grouped_df.pivot(index='Year of Birth', columns='Child\'s First Name', values='Growth Rate')

# Plot the line chart
pivot_df.plot(kind='line', legend=False)
plt.title('Growth Rate for Each Name for Each Year')
plt.xlabel('Year of Birth')
plt.ylabel('Growth Rate')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('s3://ia340-sakyitamai/Popular_Baby_Names.csv')

# Calculate the total count for each name
name_counts = df.groupby('Child\'s First Name')['Count'].sum()

# Plot the pie chart for the top 10 names
name_counts.sort_values(ascending=False).head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Names')
plt.ylabel('')
plt.show()
```
